```python
import numpy as np
import pandas as pd
import grey_functions as gf
from IPython.display import display

len_min = 3
len_max = 7

col_sdr = 'Source data range'
col_pv = '№ predict value'
col_tv = 'True value'
col_gm = 'GM'
col_mgm = 'MGM'
col_regm = 'error GM,%'
col_remgm = 'error MGM,%'

# read from input.txt file
data = np.loadtxt("input.txt", delimiter=' ', dtype=np.float)
```

```python
# N - кол-во исходных данных
# k - текущая длина промежутка для построения модели
# len_min, len_max - минимальная и максимальная длина промежутка

error_MGM = []
error_GM  = []

dfs = []
for k in range(len_min, len_max+1):
    print("k =", k,end=" ")
    order = []
    predict_mgm = []
    predict_gm = gf.grey_method_prediction(data[len_max-k:len_max], len(data)-len_max+k)[k-1:]
    index_value = [x for x in range(len_max+1, len(data)+1)]
    true_value = data[len_max:len(data)]
    for i in range(len(data)-len_max):
        order.append(str(len_max+1-k+i)+"-"+str(len_max+i))
        predict_mgm.append(gf.metabolic_grey_method_prediction(data[len_max-k+i:len_max+i],1)[0][2])
    df = pd.DataFrame({
        col_sdr: order,
        col_pv: index_value,
        col_tv: true_value,
        col_mgm: np.round(predict_mgm,3),col_remgm: np.round(np.abs(predict_mgm - true_value)/true_value*100,2),
        col_gm : np.round(predict_gm,3), col_regm: np.round(np.abs(predict_gm - true_value)/true_value*100,2)
    })
    error_MGM.append([np.round(df[col_remgm].mean(),3),k])
    error_GM.append([np.round(df[col_regm].mean(),3),k])
    display(df)
    print('Средняя относительная ошибка MGM:',np.round(df[col_remgm].mean(),3))
    print('Средняя относительная ошибка GM:', np.round(df[col_regm].mean(), 3))
    print()

error_MGM.sort()
error_GM.sort()

print("С",error_GM[0][1],"значениями у GM лучший результат со средней ошибкой",error_GM[0][0])
print("С",error_MGM[0][1],"значениями у MGM лучший результат со средней ошибкой",error_MGM[0][0])

print("\n\nВведите какую модель использовать, с каким кол-вом значений и период прогноза")

model, n, p = input("(mgm/gm) (1-{0}) (1-{1}): ".format(len(data),len(data)-1)).split()
n, p = int(n), int(p) 
while (model != ('mgm' or 'gm') or ((1 > n) and (n > len(data))) or ((1 > p) and (p > len(data)-1))):
    model, n, p = input("(mgm/gm) (1-{0}) (1-{1}): ".format(len(data),len(data)-1)).split()
    n, p = int(n), int(p)
if model == 'mgm':
    t = gf.metabolic_grey_method_prediction(data[len(data)-n:],p)
elif model == 'gm':
    t = gf.grey_method_prediction(data[len(data)-n:],p)
else:
    print("Error....")
for i in t:
    print (i)

```
